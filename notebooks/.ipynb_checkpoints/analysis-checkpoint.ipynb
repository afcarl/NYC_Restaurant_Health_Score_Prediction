```python
from scipy import stats
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [8, 5]
```

```python
data = pd.read_csv('../output/merged.csv', index_col=0)
plt.hist(data.avg_score.values,10)
plt.title('Histogram of the average health scores for NYC restaurants')
plt.savefig('../imgs/PCA_hist_1.png')
plt.show()
```

```python
components = data.columns[:-1].tolist()
X = np.array(data[components])
pca = PCA(n_components=8, whiten=True).fit(X)
X_pca = pca.transform(X)
```

```python
lmData=pd.DataFrame(columns=['P1','P2','P3','P4','P5','P6','P7','P8'])
i = 0
for col in lmData.columns:
    lmData[col] = X_pca[:,i]
    i += 1
lmData['Y'] = data['avg_score'].tolist()
lmData = lmData.reindex(np.random.permutation(lmData.index))
train = lmData[0:120]
valid = lmData[121::]
lm_pca = smf.ols(formula='Y ~ P1+P2+P3+P4+P5+P6+P7+P8',data=train).fit()
lm_pca.summary()
```

```python
r2=[lm_pca.rsquared]
lmy2 = lm_pca.predict(valid)
R2v=1-sum(pow(lmy2-valid.Y,2))/sum(pow(valid.Y-np.mean(valid.Y),2))
print('Validation R2 after PCA (unnormalized)={0}'.format(R2v))
```

```python
plt.scatter(lm_pca.predict(lmData),lmData.Y)
plt.plot(lmData.Y,lmData.Y,'r')
plt.title('Prediction of unnormalized data after PCA')
plt.xlabel('Predicted average score in each zip code')
plt.ylabel('Actual average score in each zip code')
plt.savefig('../imgs/PCA_unnorm.png')
plt.show()
```

```python
normlist = ['num_complaints', 'num_food_est', 'num_food_poi','num_smoking', 'num_violations','num_Cviolations', 
            'num_amer', 'num_chi', 'num_jap', 'num_cafe','num_ita', 'num_latin', 'num_mex', 'num_pizza', 'num_span']
```

```python
for i in normlist:
    data[i] = data[i]/data['num_restaurants']
X = np.array(data[components])
pca = PCA(n_components=8, whiten=True).fit(X)
X_pca = pca.transform(X)
```

```python
lmData=pd.DataFrame(columns=['P1','P2','P3','P4','P5','P6','P7','P8'])
i = 0
for col in lmData.columns:
    lmData[col] = X_pca[:,i]
    i += 1
lmData.index = data.index
lmData['Y'] = data['avg_score'].tolist()
lmData = lmData.reindex(np.random.permutation(lmData.index))
train = lmData[0:120]
valid = lmData[121::]
lm_pca = smf.ols(formula='Y ~ P1+P2+P3+P4+P5+P6+P7+P8',data=lmData).fit()
lm_pca.summary()
```

```python
r2=[lm_pca.rsquared]
lmy2 = lm_pca.predict(valid)
R2v=1-sum(pow(lmy2-valid.Y,2))/sum(pow(valid.Y-np.mean(valid.Y),2))
print('Validation R2 using 8 regressors after PCA (normalized)={0}'.format(R2v))
```

```python
plt.scatter(lm_pca.predict(lmData),lmData.Y)
plt.plot(lmData.Y,lmData.Y,'r')
plt.title('Prediction of normalized data after PCA (8 regs)')
plt.xlabel('Predicted average score in each zip code')
plt.ylabel('Actual average score in each zip code')
plt.savefig('../imgs/PCA_norm_8.png')
plt.show()
```

```python
lm_pca = smf.ols(formula='Y ~ P1+P2+P5+P6+P7',data=lmData).fit()
lm_pca.summary()
```

```python
r2=[lm_pca.rsquared]
lmy2 = lm_pca.predict(valid)
R2v=1-sum(pow(lmy2-valid.Y,2))/sum(pow(valid.Y-np.mean(valid.Y),2))
print('Validation R2 using 5 best regressors after PCA (normalized)={0}'.format(R2v))
```

```python
plt.scatter(lm_pca.predict(lmData),lmData.Y)
plt.plot(lmData.Y,lmData.Y,'r')
plt.title('Prediction of normalized data after PCA (5 best regs)')
plt.xlabel('Predicted average score in each zip code')
plt.ylabel('Actual average score in each zip code')
plt.savefig('../imgs/PCA_norm_5.png')
plt.show()
```

```python
lmData['Predicts'] = lm_pca.predict(lmData)
lmData['Dev_SqDist'] = (lm_pca.predict(lmData)-lmData.Y)**2
lmData.head()
```

```python
lmData.to_csv('../output/pca.csv')
```
