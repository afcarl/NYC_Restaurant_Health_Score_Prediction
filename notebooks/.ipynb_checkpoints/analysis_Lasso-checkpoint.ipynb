```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LassoCV
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = [8, 5]
```

```python
data = pd.read_csv('../output/merged.csv', index_col=0)

#Feature engineering
normList = ['num_complaints',
            'num_food_est',
            'num_food_poi',
            'num_smoking',
            'num_violations',
            'num_Cviolations',
            'num_amer',
            'num_chi',
            'num_jap',
            'num_cafe',
            'num_ita',
            'num_latin',
            'num_mex',
            'num_pizza',
            'num_span']

for i in normList:
    data[i] /= data['num_restaurants']
```

```python
data = data.reindex(np.random.permutation(data.index))
train = data[0:160]
test = data[160::]
```

```python
features = data.columns[:-1].tolist()
X,y = train[features],train['avg_score']
X_test,y_test = test[features],test['avg_score']
#Setting up Lasso cross validation
modelCV = LassoCV(alphas=np.logspace(-5, 0, num=21), normalize=True, max_iter=1e6, cv=5).fit(X, y)
print 'R2 value of the training set after CV: %.4f' % modelCV.score(X,y)
print 'R2 value of the test set using the chosen model: %.4f' % modelCV.score(X_test,y_test)
print
modelCV_weights = dict(zip(features,modelCV.coef_))
print 'Weights of each variable in our model:'
modelCV_weights
```

```python
import json
with open('../output/Lasso_weights.json', 'w') as fp:
    json.dump(modelCV_weights, fp)
```

```python
selectedFeatures = []
for i in np.nonzero(modelCV.coef_)[0]:
    selectedFeatures += [features[i]]
```

```python
X_whole,y_whole = data[features],data['avg_score']
X_preds = modelCV.predict(X_whole)
df_model = pd.DataFrame(data=data[selectedFeatures+['avg_score']], index=data.index)
df_model['Predictions'] = X_preds
df_model['Dev_SqDists'] = (X_preds-df_model['avg_score'])**2
df_model = df_model.sort_index()
df_model.to_csv('../output/Lasso.csv')
df_model.head()
```

```python
print 'R2 value of the whole set using the chosen model: %.4f' % modelCV.score(X_whole,y_whole)
```

```python
plt.scatter(modelCV.predict(X_whole),y_whole)
plt.plot(y_whole,y_whole,'r',linewidth=1.5)
plt.title('Prediction by LassoCV (7 features): R2=%.2f' % modelCV.score(X_whole,y_whole))
plt.xlabel('Predicted average score in each zip code')
plt.ylabel('Actual average score in each zip code')
plt.savefig('../figures/Lasso_7_features.png')
plt.show()
```

```python
plt.hist(df_model.Predictions.values-df_model.avg_score.values,10)
plt.title('Histogram of the deviated distances (LassoCV)')
plt.savefig('../figures/LassoCV_hist.png')
plt.show()
```
